3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
- Concept of Dispersion
Dispersion measures how spread out data points are around a central value. It helps understand variability in a dataset.
Variance & Standard Deviation
Both measure how far data deviates from the mean.
1.	Variance (σ2\sigma^2σ2)
	Average of squared differences from the mean.
	Higher variance = greater spread.
	Formula:
σ2=∑(X−μ)2N\sigma^2 = \frac{\sum (X - \mu)^2}{N}σ2=N∑(X−μ)2
2.	Standard Deviation (σ\sigmaσ)
	Square root of variance, giving dispersion in the original unit.
	Formula:
σ=σ2\sigma = \sqrt{\sigma^2}σ=σ2
	Lower σ → Data close to mean; Higher σ → Data widely spread.

4.  What is a box plot, and what can it tell you about the distribution of data?
-Box Plot (Box-and-Whisker Plot)
A box plot is a graphical representation of data distribution that shows the spread and central tendency using five key statistics:
1.	Minimum (lowest value, excluding outliers)
2.	First Quartile (Q1) – 25% of data below this point
3.	Median (Q2) – Middle value (50th percentile)
4.	Third Quartile (Q3) – 75% of data below this point
5.	Maximum (highest value, excluding outliers)
What Can a Box Plot Tell You?
 Spread of Data – The length of the box and whiskers show how spread out the data is.
 Skewness – If the median is closer to Q1 or Q3, the data is skewed.
 Outliers – Points outside the whiskers indicate extreme values.
 Symmetry – A symmetric box means evenly distributed data.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
- Concept of Skewness
Skewness measures the asymmetry of a dataset’s distribution. It indicates whether data is more concentrated on one side of the mean.
Types of Skewness
1.	Positive Skew (Right-Skewed)
o	Tail on the right, with most values concentrated on the left.
o	Mean > Median > Mode
o	Example: Income distribution (a few people earn very high salaries).
2.	Negative Skew (Left-Skewed)
o	Tail on the left, with most values on the right.
o	Mean < Median < Mode
o	Example: Exam scores (most students score high, few fail).
3.	Zero Skew (Symmetrical Distribution)
o	Data is evenly distributed around the mean.
o	Mean = Median = Mode
o	Example: Heights of a large population.
How Skewness Affects Interpretation of Data
 Affects Mean & Median → In skewed data, the mean is pulled toward the tail.
 Misleading Averages → The mean may not represent the "typical" value.
 Influences Decision-Making → Businesses analyze skewness in sales, income, and customer behavior.

7. What is the interquartile range (IQR), and how is it used to detect outliers?
- Interquartile Range (IQR) & Outlier Detection
What is IQR?
The Interquartile Range (IQR) measures the spread of the middle 50% of a dataset. It is the difference between the third quartile (Q3) and the first quartile (Q1):
IQR=Q3−Q1IQR = Q3 - Q1IQR=Q3−Q1
How is IQR Used to Detect Outliers?
Outliers are values significantly lower or higher than most of the data. The IQR method helps identify them using the following rule:
•	Lower Bound = Q1−1.5×IQRQ1 - 1.5 \times IQRQ1−1.5×IQR
•	Upper Bound = Q3+1.5×IQRQ3 + 1.5 \times IQRQ3+1.5×IQR
Any value outside these bounds is considered an outlier.

8. Discuss the conditions under which the binomial distribution is used.
- Conditions for Using the Binomial Distribution
The binomial distribution is used for modeling the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes (success or failure).

Key Conditions:
Fixed Number of Trials (n):

The experiment is repeated n times.

Example: Flipping a coin 10 times.

Two Possible Outcomes (Success or Failure):

Each trial results in either success (p) or failure (1 - p).

Example: A free throw in basketball (hit or miss).

Constant Probability (p):

The probability of success remains the same for each trial.

Example: Rolling a fair die (probability of rolling a 6 is always 1/6).

Independence of Trials:

Each trial's outcome does not affect the others.

Example: Drawing cards with replacement ensures independence.

-Normal Distribution (Properties):
•	Bell-shaped and symmetric around the mean.
•	Mean = Median = Mode.

•	Total area under the curve = 1.
•	Tails extend infinitely.
Empirical Rule (68-95-99.7 Rule):
In a normal distribution:
•	68% of data lies within 1σ of the mean.
•	95% within 2σ.
•	99.7% within 3σ.
